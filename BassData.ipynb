```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.05  # Innovation coefficient
q = 0.35  # Imitation coefficient
M = 10000  # Total potential adopters
time = np.arange(1, 13)  # Time period in months

# Bass Model Equation
def bass_model(t, p, q, M):
    exp_term = np.exp(-(p + q) * t)
    adoption = M * (1 - exp_term) / (1 + (q / p) * exp_term)
    return adoption

# Cumulative adoption over time
cumulative_adoption = bass_model(time, p, q, M)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(time, cumulative_adoption, label="Cumulative Adoption")
plt.xlabel('Time (months)')
plt.ylabel('Cumulative Adopters')
plt.title('Bass Diffusion Model - Cumulative Adoption')
plt.legend()
plt.grid(True)
plt.show()
```
